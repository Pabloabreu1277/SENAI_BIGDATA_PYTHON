# Apagar dados Fazios das colunas (axis{0 or ‘index’, 1 or ‘columns’}, default 0)

```python
futebol_df = futebol.dropna(axis=1)
futebol_df = futebol.dropna(axis=0)
display(futebol_df)
```

# Analisando os melhores resultados da espanha em todos os anos

```python
espanha_total = futebol_df.loc[futebol_df['Country'] == 'spain' ]
display(espanha_total)
espanha_total = espanha_total[['Home_Team','Home_Score','season']]
display(espanha_total)
espanha_total = espanha_total.loc[espanha_total['season'] >= 2020 ] 
display(espanha_total)
```

```python
x = espanha_total['Home_Team']
y = espanha_total['Home_Score']


plt.barh(x,y,color='red')
# plt.xlabel("Home_Score")
# plt.ylabel("Home_Team")
plt.title("Ranking de melhores times da espanha pontuando em casa 2021")
# plt.show()
```
