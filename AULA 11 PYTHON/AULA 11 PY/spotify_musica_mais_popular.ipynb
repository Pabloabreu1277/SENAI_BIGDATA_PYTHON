```python
#importando bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('./sample_data/datasetspot.csv')
display(df)
```

#Tratamento dos dados em vazio

```python
display(df.isnull().sum())
plt.figure( figsize=(15,6) )
plt.title('Analisando Campos Nulos')
sns.heatmap( df.isnull(), cbar=False );
```

##Qual artista é o mais popular da plataforma?

```python
populares = df[['artists','popularity','track_genre','track_name']]

display(populares)
```

```python
#ordenando pelo indice de popularidade do maior para o menor
popular = populares.sort_values(['popularity'], axis = 0,ascending = False)
display(popular)
```

```python
#filtrando os indices de popularidade x genero de musica
popular_genero = popular.loc[popular['popularity'] >= 95 ]
#display(populares)
sns.boxplot(data=popular_genero, x='popularity', y='track_genre')
```

```python
# Conclusão = Analisando os generos os melhores incices na media é o genero dance
```

#Analisando o genero Dance

```python
popular_dance = popular.loc[popular['track_genre'] == 'dance']
display(popular_dance)
```

```python
#Podium das musicas mais populares do spotify do genero de dance
sns.barplot(data=popular_dance, x='artists',y='popularity')
plt.show()
```

#Filtrando pelo artista

```python
#filtrando os indices de popularidade x genero de musica
popular_artista = popular.loc[popular['popularity'] >= 90 ]
sns.boxplot(data=popular_artista, x='popularity', y='artists')
```

#Analisando o artista com mais popularidade

```python
popular_sam = popular.loc[popular['artists'] == 'Sam Smith;Kim Petras']
display(popular_sam)
```
